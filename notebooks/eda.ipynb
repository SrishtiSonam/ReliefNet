```python
import pandas as pd

df_flooded_area = pd.read_csv("/content/sample_data/District_FloodedArea.csv")
df_nfhs5 = pd.read_csv("/content/sample_data/NFHS-5_HP.csv")
```

```python
print("First 5 rows of df_flooded_area:")
display(df_flooded_area.head())
print("\nData types of df_flooded_area:")
df_flooded_area.info()
print("\nDescriptive statistics for df_flooded_area:")
display(df_flooded_area.describe())

print("\nFirst 5 rows of df_nfhs5:")
display(df_nfhs5.head())
print("\nData types of df_nfhs5:")
df_nfhs5.info()
print("\nDescriptive statistics for df_nfhs5:")
display(df_nfhs5.describe())
```

```python
print("Missing values in df_flooded_area:")
display(df_flooded_area.isnull().sum())
print("\nMissing values in df_nfhs5:")
display(df_nfhs5.isnull().sum())
```

```python
# Handle missing values in df_nfhs5
# Drop 'Flag_NFHS5' and 'Flag_NFHS4' due to a high percentage of missing values.
df_nfhs5 = df_nfhs5.drop(['Flag_NFHS5', 'Flag_NFHS4'], axis=1)
# Fill missing values in 'NFHS5' and 'NFHS4' with the mean of the respective columns.
df_nfhs5['NFHS5'].fillna(df_nfhs5['NFHS5'].mean(), inplace=True)
df_nfhs5['NFHS4'].fillna(df_nfhs5['NFHS4'].mean(), inplace=True)

print("Missing values in df_nfhs5 after handling:")
display(df_nfhs5.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# 1. Histogram of 'Corrected_Percent_Flooded_Area'
fig1, ax1 = plt.subplots(figsize=(8, 6))
ax1.hist(df_flooded_area['Corrected_Percent_Flooded_Area'], bins=30, edgecolor='black')
ax1.set_title('Distribution of Corrected Percent Flooded Area')
ax1.set_xlabel('Corrected Percent Flooded Area')
ax1.set_ylabel('Frequency')
plt.tight_layout()
display(fig1)

# 2. Box plot of 'Corrected_Percent_Flooded_Area'
fig2, ax2 = plt.subplots(figsize=(8, 6))
ax2.boxplot(df_flooded_area['Corrected_Percent_Flooded_Area'])
ax2.set_title('Box Plot of Corrected Percent Flooded Area')
ax2.set_ylabel('Corrected Percent Flooded Area')
display(fig2)

# 3. Scatter plot of 'NFHS5' vs 'NFHS4'
fig3, ax3 = plt.subplots(figsize=(8, 6))
ax3.scatter(df_nfhs5['NFHS4'], df_nfhs5['NFHS5'], alpha=0.5)
ax3.set_title('NFHS5 vs NFHS4')
ax3.set_xlabel('NFHS4')
ax3.set_ylabel('NFHS5')
display(fig3)

# 4. Histogram of 'NFHS5'
fig4, ax4 = plt.subplots(figsize=(8, 6))
ax4.hist(df_nfhs5['NFHS5'], bins=30, edgecolor='black')
ax4.set_title('Distribution of NFHS5')
ax4.set_xlabel('NFHS5')
ax4.set_ylabel('Frequency')
plt.tight_layout()
display(fig4)

# 5. Box plot of 'NFHS5'
fig5, ax5 = plt.subplots(figsize=(8, 6))
ax5.boxplot(df_nfhs5['NFHS5'])
ax5.set_title('Box Plot of NFHS5')
ax5.set_ylabel('NFHS5')
display(fig5)

plt.close('all') # Close figures to prevent them from displaying twice
```

```python
# 1. Calculate and display mean, median, and standard deviation
mean_flooded = df_flooded_area['Corrected_Percent_Flooded_Area'].mean()
median_flooded = df_flooded_area['Corrected_Percent_Flooded_Area'].median()
std_flooded = df_flooded_area['Corrected_Percent_Flooded_Area'].std()

print(F"Mean Corrected Percent Flooded Area: {mean_flooded:.2f}")
print(F"Median Corrected Percent Flooded Area: {median_flooded:.2f}")
print(F"Standard Deviation of Corrected Percent Flooded Area: {std_flooded:.2f}")

# 2. Identify and display districts with highest and lowest 'Corrected_Percent_Flooded_Area'
highest_flooded_district = df_flooded_area.loc[df_flooded_area['Corrected_Percent_Flooded_Area'].idxmax()]
lowest_flooded_district = df_flooded_area.loc[df_flooded_area['Corrected_Percent_Flooded_Area'].idxmin()]

print("\nDistrict with Highest Corrected Percent Flooded Area:")
display(highest_flooded_district)

print("\nDistrict with Lowest Corrected Percent Flooded Area:")
display(lowest_flooded_district)

# 3. Analyze the relationship between 'Percent_Flooded_Area' and 'Parmanent_Water'
plt.figure(figsize=(8, 6))
plt.scatter(df_flooded_area['Percent_Flooded_Area'], df_flooded_area['Parmanent_Water'], alpha=0.5)
plt.title('Relationship between Percent Flooded Area and Permanent Water')
plt.xlabel('Percent Flooded Area')
plt.ylabel('Permanent Water')
plt.grid(True)
plt.show()

correlation = df_flooded_area['Percent_Flooded_Area'].corr(df_flooded_area['Parmanent_Water'])
print(F"\nCorrelation coefficient between Percent Flooded Area and Permanent Water: {correlation:.2f}")
```

```python
# 1. Get a list of unique indicators
unique_indicators = df_nfhs5['Indicator'].unique()
print("Unique Indicators:")
display(unique_indicators)

# 2. Select a few key indicators of interest
selected_indicators = [
    'Female population age 6 years and above who ever attended school (%)',
    'Population below age 15 years (%)',
    'Sex ratio of the total population (females per 1,000 males)',
    'Children under age 5 years whose birth was registered (%)',
    'Women age 15-49 years who are anemic (%)',
    'Men age 15-49 years who are anemic (%)',
    'Households with access to an improved sanitation facility (%)',
    'Population with a usual source of drinking water located on premises (%)'
]
print("\nSelected Key Indicators:")
display(selected_indicators)

# 3. & 4. Filter, calculate, and display mean and std dev for each selected indicator
print("\nMean and Standard Deviation for Selected Indicators:")
for indicator in selected_indicators:
    df_indicator = df_nfhs5[df_nfhs5['Indicator'] == indicator]
    mean_nfhs5 = df_indicator['NFHS5'].mean()
    std_nfhs5 = df_indicator['NFHS5'].std()
    mean_nfhs4 = df_indicator['NFHS4'].mean()
    std_nfhs4 = df_indicator['NFHS4'].std()

    print(F"\nIndicator: {indicator}")
    print(F"  NFHS5 - Mean: {mean_nfhs5:.2f}, Std Dev: {std_nfhs5:.2f}")
    print(F"  NFHS4 - Mean: {mean_nfhs4:.2f}, Std Dev: {std_nfhs4:.2f}")

# 5. Create visualizations
print("\nVisualizations for Selected Indicators:")
for indicator in selected_indicators:
    df_indicator = df_nfhs5[df_nfhs5['Indicator'] == indicator]

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(F'Distribution of NFHS5 and NFHS4 for "{indicator}"', y=1.02)

    # Box plot for NFHS5
    axes[0].boxplot(df_indicator['NFHS5'].dropna())
    axes[0].set_title('NFHS5')
    axes[0].set_ylabel('Value')
    axes[0].set_xticks([]) # Hide x-axis ticks for box plot

    # Box plot for NFHS4
    axes[1].boxplot(df_indicator['NFHS4'].dropna())
    axes[1].set_title('NFHS4')
    axes[1].set_ylabel('Value')
    axes[1].set_xticks([]) # Hide x-axis ticks for box plot

    plt.tight_layout()
    display(fig)
    plt.close(fig)
```
